```python
import pdfplumber
import os

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

def save_text_to_file(text, output_folder, pdf_file_name):
    os.makedirs(output_folder, exist_ok=True)
    text_file_path = os.path.join(output_folder, f"{os.path.splitext(pdf_file_name)[0]}.txt")
    
    if os.path.exists(text_file_path):
        print(f"Text file already exists: {text_file_path}, skipping...\n")
        return
    
    with open(text_file_path, "w", encoding="utf-8") as file:
        file.write(text)
    print(f"Saved extracted text: {text_file_path}")

pdf_folder = os.path.expanduser("~/Downloads/pdf_cases")
pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith(".pdf")]
output_folder = os.path.join(pdf_folder, "extracted_texts")

for pdf_file in pdf_files:
    pdf_path = os.path.join(pdf_folder, pdf_file)
    text_file_path = os.path.join(output_folder, f"{os.path.splitext(pdf_file)[0]}.txt")
    
    if not os.path.exists(text_file_path):
        extracted_text = extract_text_from_pdf(pdf_path)
        save_text_to_file(extracted_text, output_folder, pdf_file)
    else:
        print(f"Extracted text file already exists for {pdf_file}, skipping...\n")
```
